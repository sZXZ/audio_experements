```python
# n_fft (like 2048 or 4096) for constant hums vs. lower values for sharper sounds.
folder_with_samples = Path('/Users/alex/Music/Samples/scissors')
base_file_name = 's_3_cutp'
audio = folder_with_samples.joinpath(f'{base_file_name}.wav')
noise = folder_with_samples.joinpath(f'{base_file_name}_noise.wav')
n_fft_setting = 1024
# denoising outs
out1 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5.wav')
remove_noise(audio, output_path=out1, prop_decrease=0.5)

out2 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_twice.wav')
remove_noise(out1, output_path=out2, prop_decrease=0.5)

out3 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_twice_1.wav')
remove_noise(out2, output_path=out3, prop_decrease=1)

out4 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_1.wav')
remove_noise(audio, output_path=out4, prop_decrease=1)

out5 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_1_wn.wav')
remove_noise(audio, output_path=out5, noise_path=noise, prop_decrease=1)

out6 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_wn.wav')
remove_noise(audio, output_path=out6, noise_path=noise, prop_decrease=0.5)

out7 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_wn_twice.wav')
remove_noise(out6, output_path=out7, noise_path=noise, prop_decrease=0.5)

out8 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_wn_twice_1_wn.wav')
remove_noise(out7, output_path=out8, noise_path=noise, prop_decrease=1)

out9 = folder_with_samples.joinpath(f'{base_file_name}_dn_py_0_5_wn_twice_1.wav')
remove_noise(out7, output_path=out9, prop_decrease=1)
```
